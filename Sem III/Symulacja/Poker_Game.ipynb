```python
import pandas as pd
from Poker_Sim import simulation
from Poker_Player import AgressivePlayer, PassivePlayer, ConsistentPlayer, PatientPlayer, AveragePlayer
from Poker_Params import df_params
```

```python

df = df_params
```

```python
player_types = [ConsistentPlayer, PatientPlayer, PassivePlayer, AgressivePlayer, AveragePlayer]

for player_type in player_types:
    df["Typ gracza"] = df["Typ gracza"].apply(lambda x: player_type if x == player_type.__name__ else x)
```

```python
columns = ["Typ gracza", "Pieniądze na wstępie", "Lekkomyślność", "Rozrzutność", "Minimalna oczekiwana ręka"] + [f"Wynik_pieniężny_gra_{i+1}" for i in range(10)]
wyniki = pd.DataFrame(columns=columns)

o = 0
l = len(df.values)*3
for params in df.values:
    
    for i in range(3):
        print(o,'/',l)
        wyniki.loc[len(wyniki)] = simulation(*params)
        o += 1
    
```

```python
file_path = "ref_wyniki_symulacji_poker.csv"
wyniki.to_csv(file_path, index=False)
```
